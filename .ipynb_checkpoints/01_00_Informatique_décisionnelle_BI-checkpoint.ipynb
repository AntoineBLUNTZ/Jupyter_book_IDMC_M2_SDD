##### **Période Couverte**
- **Années** : 2013, 2014 (estimé)
- **Prévisions** : 2015, 2016

##### **Variables/Indicateurs**
Pour chaque année, trois indicateurs répétés :
1. **Recettes totales avec dons** (% du PIB)
2. **Dépenses totales et prêts nets** (% du PIB)
3. **Solde global** (% du PIB) = Recettes - Dépenses

##### **Sources Officielles**
- Département des statistiques BafD
- FMI (Fonds Monétaire International)
- Diverses autorités nationales
- Estimations (2014) et prévisions (2015-2016) du Département des statistiques BafD

##### **Notes Spéciales**
- Astérisque (*) : Année fiscale Juillet (n-1)/Juin (n) - Égypte
- Astérisques doubles (**) : Année fiscale avril (n)/mars (n+1) - Botswana, Lesotho, Namibie, Swaziland
- Valeurs manquantes : Somalie (tout), Soudan du Sud (tout sauf les soldes globaux)

##### **Indicateurs Clés**
- **Solde global positif** : Surplus budgétaire (rare)
- **Solde global négatif** : Déficit budgétaire (courant)
- Comparaison recettes/dépenses : mesure de discipline fiscale

##### **Utilisations Principales**
- Analyse de la soutenabilité budgétaire
- Suivi de l'exécution fiscale
- Identification des pays à risque de crise
- Prévisions de besoins de financement
- Évalation de politique budgétaire
