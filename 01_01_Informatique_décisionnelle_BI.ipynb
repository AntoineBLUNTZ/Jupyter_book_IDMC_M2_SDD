##### **Période Couverte**
- **Année de référence principale** : 2014
- **Taux de croissance du PIB** : Moyenne annuelle sur 2006-2014

##### **Variables/Indicateurs**
- **Population** (en milliers d'habitants)
- **Superficie** (en milliers de km²)
- **Densité de population** (population par km²)
- **Produit intérieur brut** (PIB) (en millions de dollars en parité de pouvoir d’achat (PPA))
- **PIB par habitant** (en PPA)
- **Taux de croissance du PIB en volume** (moyenne annuelle 2006-2014, %)

<u>NB:</u> La mesure du PIB en PPA permet comparer la richesse réelle des habitants entre différents pays, en tenant compte des différences de niveaux de prix et de pouvoir d'achat, plutôt que de se baser uniquement sur les taux de change de change de marché.

##### **Sources Officielles**
- **Population** : Nations Unies, Division de la population, Département des Affaires économiques et sociales, World Population Prospects, The 2012 Revision.
- Département des statistiques de la Banque Africaine de Développement (BafD)
- Diverses autorités nationales
- Estimations et prévisions du Département des statistiques BafD

##### **Notes Spéciales**
- Pour l'Égypte : année fiscale juillet (n-1)/juin (n) -> pas utile ?
- Valeurs manquantes : Somalie (PIB, PIB par hab, Taux de croissance du PIB), Soudan du Sud (Taux de croissance du PIB)

<u>NB:</u> Explications possibles des valeurs manquantes : instabilité politique et conflits prolongés, manque d'infrastructures pour la collecte exhaustive et régulière des données.

##### **Utilisations Principales**
- Comparaisons démographiques
- Calcul d'agrégats économiques (voir les CER)
- Comparaisons de PIB par habitant
- Analyse de densité de population
- Analyse des données de croissance (évaluer l'efficacité des politiques de développement)
