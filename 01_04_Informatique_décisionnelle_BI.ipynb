##### **Période Couverte**
- Année de référence principale : 2013
- Données détaillées : 2014 (estimé- e), 2015 (prévision - p), 2016 (prévision - p)

##### **Variables/Indicateurs**
**Section 1 : Dette Extérieure Totale**
- **Total en 2013** (en millions USD)

**Section 1b : Composition de la Dette (pourcentage du total)**

Pour 2013 :
1. Multilatérale (%) : ensemble des dettes contractées par un État auprès d'un groupe d'États ou d'une institution financière internationale (FMI, Banque mondiale ...)
2. Bilatérale (%) : ensemble des engagements contractés par un État auprès d'un autre État
3. Privée (%) : ensemble des dettes contractées par l'ensemble des agents économiques privés du pays : ménages, entreprises, artisans, commerçants, professions libérales, agriculteurs, etc.

**Section 2 : Dette Extérieure Totale (% du PIB)**
- 2013, 2014 (e), 2015 (p), 2016 (p)

**Section 3 : Service de la Dette (% des exportations)**
- Montant des paiements dus pour une année, indicateur clé du fardeau de la dette d’un pays
- 2013, 2014 (e), 2015 (p), 2016 (p)

##### **Indicateurs Clés**
- **Endettement soutenable** : <60% du PIB (critère de Maastricht)
- **Risque modéré** : 30-60% du PIB
- **Seuil de viabilité de la dette** : entre 15 et 25 % des exportations selon le FMI et la Banque mondiale (essayer de voir avec les pays avec >20% ?)

##### **Utilisations Principales**
- Analyse de soutenabilité de la dette
- Identification des pays à risque de crise
- Suivi de la composition des créanciers
- Prévisions de besoins de refinancement
- Évaluation de capacité de remboursement
