##### **Notes Spéciales**
- Les taux d'analphabétisme représentent le complément : 100 % - % alphabétisés
- Les dépenses incluent généralement l'enseignement primaire, secondaire et supérieur
- Valeurs manquantes : Taux d'analphabétisme de Djibouti, Somalie et Soudan du Sud + dépenses publique d'éducation en Somalie (à 0 mais ne peut pas être calculé si PIB non connu). Causes : instabilité politique et conflits prolongés, manque d'infrastructures pour la collecte exhaustive et régulière des données, volonté de cacher ces données.
- Valeurs à expliquer : les dépenses publiques d'éducation pour la Guinée-Bissau, la Lybie et le Nigéria sont nulles et les taux estimés d'analphabétisme de 99.9% en Lybie pour les 15-24 ans.

##### **Indicateurs Clés**
- **Taux d'alphabétisation mondial en 2015** : hommes: 90,0 %, femmes : 82,7 %, total : 86,3 % (CIA World Factbook)
- **Disparités entre les sexes** : Écart (hommes - femmes) en points de pourcentage
- **Efficacité des dépenses** : Ratio dépenses/résultats d'alphabétisation
