##### **Période Couverte**
- **Emploi** : 2012
- **Remises de travailleurs** : 2010, 2011, 2012, 2013, 2014 (estimé - e)

##### **Variables/Indicateurs**
**Section 1 : Taux de Chômage (âge 15-64)**
- **Année de base** : 2012 (données harmonisées estimées par l'OIT)
- **Indicateurs** : Total, Hommes, Femmes (%)
- **Calcul** : nombre de chômeurs/population active

**Section 2 : Taux de Participation**
- **Année de base** : 2012
- **Indicateurs** : Total (âge >15), jeunes (15-24 ans) (%)
- **Interprétation** : % de population en âge de travailler qui participe activement au marché du travail, soit en travaillant, soit en cherchant activement un emploi. = % d'actifs sur la tranche d'âge considérée

**Section 3 : Taux d'Inactivité (âge 15-64)**
- **Année de base** : 2012
- **Indicateurs** : Total, Hommes, Femmes (%)
- **Interprétation** : % de population en âge de travailler qui ne participe pas activement au marché du travail, granularité par sexe

**Section 4 : Remises de Travailleurs**
- **Années** : 2010, 2011, 2012, 2013, 2014 (e)
- **Unité** : Millions de dollars USD
- **Remises de travailleurs** : également appelées envois de fonds, rémitances ou remises migratoires, désignent la part des revenus gagnés à l'étranger que les migrants rapatrient dans leur pays d'origine. -> serait intéressant de coupler avec le % de la population vivant en diaspora pour chaque pays.

##### **Sources Officielles**
- **Emploi** : 
    - Organisation internationale du Travail (OIT), base de donnée KILM, 8e édition, Trends Estimation Model
    - Données fournies par autorités nationales pour Maurice, Maroc, Namibie et Afrique du Sud
- **Remises** : Banque mondiale, Indicateurs du développement dans le monde, base de données GDF en ligne (consultée avril 2015)

##### **Indicateurs Clés**
- **Taux de chômage élevé** : >10% (problème social majeur)
- **Dépendance aux remises** : >5% du PIB indique vulnérabilité
- **Écart de participation entre les sexes** : Disparité hommes-femmes en points de pourcentage

##### **Utilisations Principales**
- Suivi du marché du travail
- Analyse des disparités de genre emploi
- Évaluation de la migration économique
- Mesure de dépendance à la diaspora
- Prévisions de besoins de création d'emplois
