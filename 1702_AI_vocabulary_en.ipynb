## **Prologue - The 4 Basic Concepts**

Before the climb, the rules of the game

### `Concept 1`&nbsp; **LLM** <span style="color:grey;font-size: 80%">‚Äî Large Language Model</span>

You are not using a "chatbot". You are using a **LLM** ‚Äî Large Language Model.

<span style="color:green">**This is**</span></br>
A linguistic probability prediction engine

<span style="color:orangered">**This is not**</span></br>
A knowledge base (like Wikipedia)

**Example** </br>
<span style="color:grey">He doesn't "know" that the sky is blue. He calculates that "blue" has a 99% chance of coming after "sky".</span>


### `Concept 2`&nbsp; **Stochastic nature** <span style="color:grey;font-size: 80%">‚Äî Throwing the dice</span>

**Calculator**</br>
$`1+1=2`$</br>
<span style="color:grey;font-size: 80%">Always. Deterministic.</span>

**LLM**</br>
üé≤üé≤üé≤</br>
<span style="color:grey;font-size: 80%">Roll dice for each word. Probabilistic.</span>

The same question can give two different answers. That's not a bug, it's its inherent nature.

<span style="font-size: 80%">That's what makes her **creative**..</span></br>
<span style="font-size: 80%">This is what makes it **non-deterministic**.</span>

### `Concept 3`&nbsp; **Inference vs Training**

<span style="color:darkorange">**Training**</span></br>
<span style="font-size: 120%">the school</span></br>
<span style="color:grey">It takes months.</span></br>
<span style="color:grey;font-size: 80%">The model learns from billions of texts. Its knowledge is defined here.</span>


<span style="color:royalblue">**Inference**</span></br>
<span style="font-size: 120%">the examination</span></br>
<span style="color:grey">It takes a second.</span></br>
<span style="color:grey;font-size: 80%">When you talk to ChatGPT, it's in inference mode. It uses what it has learned.</span>

**Important:** It doesn't learn from you in real time. Its knowledge is fixed at the end of its training. It's an encyclopedia that doesn't update itself.

### `Concept 4`&nbsp; **Transformer** <span style="color:grey;font-size: 80%">‚Äî The architecture (2017)</span>

It's the engine under the hood. The invention that changed everything.

<span style="color:grey;font-size: 80%">BEFORE (2017)</span></br>Models read word for word, sequentially.

<span style="font-size: 80%">AFTER (Transformer)</span></br>
Reading the entire context in parallel.</br>
<span style="font-size: 80%">"[Attention Mechanism](https://en.wikipedia.org/wiki/Attention_(machine_learning))"</span>

Without this invention:</br>
<span style="color:grey;font-size: 80%">‚ùå No GPT</span></br>
<span style="color:grey;font-size: 80%">‚ùå No Claude</span></br>
<span style="color:grey;font-size: 80%">‚ùå No modern AI</span></br>

### ![](media/images/icons/royal-blue_1-filled-24.ico) &nbsp;**Tokenization** <span style="font-size: 80%"><span> <span style="color:grey;font-size: 100%">‚Äî AI's atoms</span>

<span style="color:royalblue">What you see</span></br>

<span style="color:royalblue">What the AI ‚Äã‚Äãsees</span></br>
`"under"` `"stand"` `"ing"`


<span style="color:royalblue">Why this is crucial</span></br>
&nbsp; &nbsp; <span style="font-size: 90%">You're charged per token, not per word</span></br>
&nbsp; &nbsp; <span style="font-size: 90%">Explains the errors in wordplay</span></br>
&nbsp; &nbsp; <span style="font-size: 90%">The foundation of everything else</span></br>

<span style="color:grey;font-size: 80%">AI breaks the text into pieces (tokens), not into letters.</br>That is why it struggles with anagrams and letter counting.</span>

### ![](media/images/icons/royal-blue_2-filled-24.ico) &nbsp;**Embedding** <span style="color:grey;font-size: 80%">‚Äî Coordinates in the space of meaning</span>

Each token is assigned mathematical coordinates in a space with hundreds of dimensions.

Words that are similar in meaning are similar in coordinates:</br></br>
&nbsp; &nbsp; <span style="color:grey;font-size: 80%">"dog" is similar to "cat"</span></br>
&nbsp; &nbsp; <span style="color:grey;font-size: 80%">"car" is similar to "truck"</span></br>
&nbsp; &nbsp; <span style="color:grey;font-size: 80%">"democracy" is very far from "sandwich"</span></br>

<span style="color:royalblue">The famous equation</span></br>
&nbsp; <span style="font-size: 140%">king - man + woman = queen</span></br>
<span style="color:grey;font-size: 80%">AI performs this calculation in the coordinate space. It is the geometry of meaning.</span>


### ![](media/images/icons/royal-blue_3-filled-24.ico) &nbsp;**Latent space** <span style="color:grey;font-size: 80%">‚Äî AI's inner universe</span>

A cosmos where each point represents a possible idea. When you ask a question, the AI ‚Äã‚Äãtraces a path from your question to the answer.

<span style="color:orangered">Origin of hallucinations</span></br>
<span style="color:grey;font-size: 80%">When AI lands in a little-explored area, it confidently describes things that don't exist. Like a tourist giving directions in a city they've never visited.</span>

<span style="color:royalblue">Navigating the cosmos</br>of meaning</span>




### ![](media/images/icons/royal-blue_4-filled-24.ico) &nbsp;**Positional encoding** <span style="color:grey;font-size: 80%">‚Äî Words' rythm</span>

The AI ‚Äã‚Äãsees all the tokens at once, without any natural concept of 1st, 2nd, or 3rd word. A "musical signature" is added to each position.

"The cat eats the mouse"</br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="color:grey;font-size: 80%">‚â†</span></br> 
"The mouse eats the cat"

<span style="color:royalblue">It helps to untangle</span></br>
<span style="color:grey;font-size: 80%">"The report that the director who was hired last year wrote was excellent."</span></br>
<span style="color:royalblue;font-size: 80%">Who wrote what? Positional encoding keeps track.</span>

### ![](media/images/icons/green_5-filled-24.ico) &nbsp;**Prompt engineering**  <span style="color:grey;font-size: 80%">‚Äî The art of asking well</span>

<span style="color:orangered">Weak prompt</span></br>
"Write me a text about dogs"</br>
<span style="color:grey;font-size: 80%">&#8594; Vague, unusable result</span>

<span style="color:green">Strong prompt</span></br>
"Write a 200-word guide for people adopting their first dog. Focus on the first week. Include common mistakes and essential supplies. Use an encouraging tone."</br>
<span style="color:grey;font-size: 80%">&#8594; Readily actionable result</span>

**Precision = Quality**. AI is not telepathic. The more specific you are about the context, format, and tone, the more usable the result will be.

### ![](media/images/icons/green_6-filled-24.ico) &nbsp;**Temperature** <span style="color:grey;font-size: 80%">‚Äî The creativity slider</span>

<span style="color:royalblue">Low temperature</span></br>
<span style="color:grey;font-size: 80%">(close to 0)</span></br>
"The sky is blue."</br>
<span style="color:grey;font-size: 80%">Predictable, consistent, sometimes boring</span>

<span style="color:green">Average temperature</span></br>
<span style="color:grey;font-size: 80%">(~0.7)</span></br>
"The sky is cloudy today."</br>
<span style="color:grey;font-size: 80%">Balanced</span>

<span style="color:orange">High temperature</span></br>
<span style="color:grey;font-size: 80%">(close to 1+)</span></br>
"The sky is melting into purple dreams."</br>
<span style="color:grey;font-size: 80%">Creative, surprising, sometimes inconsistent</span>

**Coding, analyses, factual documents** &#8594; Low</br>
**Brainstorming, creative writing** &#8594; High

### ![](media/images/icons/green_7-filled-24.ico) &nbsp;**Context window** <span style="color:grey;font-size: 80%">‚Äî Working memory</span>

Each model has a limited capacity. Once the memory is full, something must be removed to make room.

<span style="color:orange">Amnesia explained</span></br>
<span style="color:grey;font-size: 80%">"What was the first city that we mentioned?"</br>‚Äî It might forget it. The information has slipped out of its working memory.</span>

<span style="color:green">This is not a bug</span></br>
This is an architectural limitation that you can circumvent: summarize regularly, break down the work, and remind the essential context.

GPT-4 : 128K tokens</br>
Claude : 200K tokens</br>
Gemini : 1M+ tokens

### ![](media/images/icons/green_8-filled-24.ico) &nbsp;**Sampling methods** <span style="color:grey;font-size: 80%">‚Äî Choosing the next word</span>

<span style="color:grey;font-size: 80%">For "today's weather is..." :</span>

<span style="color:royalblue">Beam search</span></br>
"... forecast to remain cloudy with occasional showers."</br>
<span style="color:grey;font-size: 80%">Explore several paths, keep the best overall path. &#8594; Methodical, meticulous.</span>

<span style="color:green">Top-K</span></br>
"... beautiful and sunny."</br>
<span style="color:grey;font-size: 80%">Consider the K most probable words. &#8594; Balance between consistency and variety.</span>

<span style="color:orange">Top-P (nuclus sampling)</span></br>
"... absolutely bizarre, it's snowing in July."</br>
<span style="color:grey;font-size: 80%">Covers 90% of the probability. &#8594; Dynamic, unpredictable.</span>

These methods create **different personalities**: meticulous editor, reliable assistant, or creative partner.
