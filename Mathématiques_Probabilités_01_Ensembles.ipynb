# **Introduction**

On a affaire à des événements qui sont aléatoires. Nous allons chercher à faire des prévisions sur ces événements. On s'entraînera sur des cas simples, avec des objets du quotidien :
- pièces de monnaie (pile ou face)
- dés
- jeu de cartes
- ...

## **Probabilités : intuitives**

Les probabilités sont d'abord intuitives. On considère une expérience pour laquelle plusieurs issues sont possibles. On peut se demander quelle est la probabilité de chacune des issues.<br>
Dans des cas simples, cela va se déterminer relativement facilement.

|  <span style="color:cornflowerblue">Expérience</span>  	| <span style="color:green">Événement</span> 	| <span style="color:green">Probabilité</span>	|
|:----------------:	|:-------------:	|:---------------:	|
| <span style="color:cornflowerblue">lancer une pièce</span> 	|  <span style="color:green">obtenir Face</span> 	|       <span style="color:green">1/2</span>       	|

Une chance sur deux : ok, très bien. On peut déjà se poser la question de ce que cela signifie et à quoi cela nous sert. Que veut dire une chance sur deux ? Est-ce que cela va nous aider à deviner quel sera le résultat du prochain lancer de la pièce ? Bien sûr que non. La probabilité n'est pas là pour ça. La probabilité n'est pas là pour nous dire ce qu'il va se passer la prochaine fois que l'on lancer une pièce.

On découvrira à quoi servent vraiment les probabilités dans le chapitre sur les lois de probabilité. Savoir que quelque chose à une chance sur deux d'arriver, cela n'est absolument pas utile si on veut essayer l'expérience une fois. Que ce soit une probabilité d'une chance sur mille (1/1000), d'une chance sur deux (1/2) ou de neuf cent quatre-vingt-dix-neuf sur mille (999/1000), ça ne nous assurera pas du résultat futur.

Continuons dans nos exemples, et demandons nous si systématiquement les probabilités peuvent être trouvées par intuition, avec simplement un peu de bon sens.

|  <span style="color:cornflowerblue">Expérience</span>  	| <span style="color:green">Événement</span> 	| <span style="color:green">Probabilité</span>	|
|:----------------:	|:-------------:	|:---------------:	|
| <span style="color:cornflowerblue">lancer une pièce</span> 	|  <span style="color:green">obtenir Face</span> 	|       <span style="color:green">1/2</span>       	|
| <span style="color:cornflowerblue">lancer un dé</span> 	|  <span style="color:green">obtenir le 1</span> 	|       <span style="color:green">1/6</span>       	|
|  	|  <span style="color:green">c'est pair</span> 	|       <span style="color:green">3/6 = 1/2</span>       	|

## **Probabilités : intuitives... ?**

Compliquons progressivement les expériences et les événements envisagés, et continuons de nous demander si les probabilités sont aussi intuitives qu'on pourrait le croire.

|  <span style="color:cornflowerblue">Expérience</span>  	| <span style="color:green">Événement</span> 	| <span style="color:green">Probabilité</span>	|
|:----------------:	|:-------------:	|:---------------:	|
| <span style="color:cornflowerblue">lancer une pièce</span> 	|  <span style="color:green">obtenir Face</span> 	|       <span style="color:green">1/2</span>       	|
| <span style="color:cornflowerblue">lancer un dé</span> 	|  <span style="color:green">obtenir le 1</span> 	|       <span style="color:green">1/6</span>       	|
|  	|  <span style="color:green">c'est pair</span> 	|       <span style="color:green">3/6 = 1/2</span>       	|
| <span style="color:cornflowerblue">lancer deux pièces</span> 	|  <span style="color:green">deux Face</span> 	|       <span style="color:green">1/4</span>
| <span style="color:cornflowerblue">lancer deux dés</span> 	|  <span style="color:green">on a un double</span> 	|       <span style="color:green">6/36 = 1/6</span>       	|
| <span style="color:cornflowerblue">tirer 5 cartes<br>d'un jeu de 32</span> 	|  <span style="color:green">avoir un brelan</span> 	|       <span style="color:green">12096/201376<br>= 54/899 $\approx$ 0,06</span>       	|  

À partir d'un certain degré de complexité, notre intuition n'est plus utilisable. On ne peut pas se fier à son bon sens si l'on souhaite faire des calculs de probabilité dans des situations à peu près quelconques.

Il va nous falloir des bases, des outils de calcul, etc. Bref, il va falloir qu'on s'organise.

Si l'on veut calculer des probabilités sur des événements donnés, il faut être capable de dénombrer des situations qui nous intéressent parmi tout un ensemble de situations possibles. Il faut donc compter ces situations. Et pour les compter, il faut savoir les classifier, c'est-à-dire les ranger dans des ensembles et des sous-ensembles.

```python
# https://stackoverflow.com/questions/5837572/generate-a-random-point-within-a-circle-uniformly/50746409#50746409

import matplotlib
from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111)
circle = matplotlib.patches.Circle((0, 0), radius=1, facecolor='yellow', alpha=0.3, edgecolor='green', linewidth=5, linestyle='solid')
ax.add_patch(circle)
#plt.text(-.4, 0, "This is a Circle")

rho = np.sqrt(np.random.uniform(0, 1, 6))
phi = np.random.uniform(0, 2*np.pi, 6)

x = rho * np.cos(phi)
y = rho * np.sin(phi)

n = ['a', 'b', 'c', 'd', 'e', 'f']

# Entire aplhabet
#n = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j',
#     'k', 'l','m', 'n', 'o', 'p', 'q', 'r', 's', 't',
#     'u', 'v', 'w', 'x', 'y', 'z']


plt.scatter(x, y, s = 0)

for i, txt in enumerate(n):
    plt.annotate(txt, (x[i], y[i]))

plt.axis('equal')

# Annotate with text + Arrow
plt.annotate(
# Label and coordinate
'Venn diagram', xy=(-1, 0), xytext=(-2, 0.2), fontsize=10,
# Custom arrow
arrowprops=dict(facecolor='black', shrink=0, alpha=0))

# Annotate with text + Arrow
plt.annotate(
# Label and coordinate
'A', xy=(0.75, 0.75), xytext=(1, 1), fontsize=10,
# Custom arrow
arrowprops=dict(facecolor='green', shrink=0, alpha=0.3))

# turn off the axis
plt.axis('off')
```

```python
# https://stackoverflow.com/questions/5837572/generate-a-random-point-within-a-circle-uniformly/50746409#50746409

from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111)
circle = matplotlib.patches.Circle((0, 0), radius=1, facecolor='yellow', alpha=0.3, edgecolor='green', linewidth=5, linestyle='solid')
ax.add_patch(circle)
#plt.text(-.4, 0, "This is a Circle")

rho = np.sqrt(np.random.uniform(0, 1, 6))
phi = np.random.uniform(0, 2*np.pi, 6)

x = rho * np.cos(phi)
y = rho * np.sin(phi)

n = ['a', 'b', 'c', 'd', 'e', 'f']

# Entire aplhabet
#n = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j',
#     'k', 'l','m', 'n', 'o', 'p', 'q', 'r', 's', 't',
#     'u', 'v', 'w', 'x', 'y', 'z']


plt.scatter(x, y, s = 0)

for i, txt in enumerate(n):
    plt.annotate(txt, (x[i], y[i]))

plt.axis('equal')

# Annotate with text + Arrow
plt.annotate(
# Label and coordinate
'Venn diagram', xy=(-1, 0), xytext=(-2, 0.2), fontsize=10,
# Custom arrow
arrowprops=dict(facecolor='black', shrink=0, alpha=0))

# Annotate with text + Arrow
plt.annotate(
# Label and coordinate
'A', xy=(0.75, 0.75), xytext=(1, 1), fontsize=10,
# Custom arrow
arrowprops=dict(facecolor='green', shrink=0, alpha=0.3))

# turn off the axis
plt.axis('off')
```

```python
# libraries
from matplotlib import pyplot as plt
import numpy as np
from matplotlib_venn import venn3, venn3_circles
 
# Make a Basic Venn
v = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels = ('A', 'B', 'C'))
 
# Custom it
v.get_patch_by_id('100').set_alpha(1.0)
v.get_patch_by_id('100').set_color('white')
v.get_label_by_id('100').set_text('Unknown')
v.get_label_by_id('A').set_text('Set "A"')
c = venn3_circles(subsets=(1, 1, 1, 1, 1, 1, 1), linestyle='dashed')
c[0].set_lw(1.0)
c[0].set_ls('dotted')
 
# Add title and annotation
plt.title("Sample Venn diagram")
plt.annotate('Unknown set', xy=v.get_label_by_id('100').get_position() - np.array([0, 0.05]), xytext=(-70,-70),
ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray'))
 
# Show it
plt.show()
```
