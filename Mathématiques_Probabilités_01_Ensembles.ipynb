# **Introduction**

On a affaire à des événements qui sont aléatoires. Nous allons chercher à faire des prévisions sur ces événements. On s'entraînera sur des cas simples, avec des objets du quotidien :
- pièces de monnaie (pile ou face)
- dés
- jeu de cartes
- ...

## **Probabilités : intuitives**

Les probabilités sont d'abord intuitives. On considère une expérience pour laquelle plusieurs issues sont possibles. On peut se demander quelle est la probabilité de chacune des issues.<br>
Dans des cas simples, cela va se déterminer relativement facilement.

|  <span style="color:cornflowerblue">Expérience</span>  	| <span style="color:green">Événement</span> 	| <span style="color:green">Probabilité</span>	|
|:----------------:	|:-------------:	|:---------------:	|
| <span style="color:cornflowerblue">lancer une pièce</span> 	|  <span style="color:green">obtenir face</span> 	|       <span style="color:green">1/2</span>       	|

Une chance sur deux : ok, très bien. On peut déjà se poser la question de ce que cela signifie et à quoi cela nous sert. Que veut dire une chance sur deux ? Est-ce que cela va nous aider à deviner quel sera le résultat du prochain lancer de la pièce ? Bien sûr que non. La probabilité n'est pas là pour ça. La probabilité n'est pas là pour nous dire ce qu'il va se passer la prochaine fois que l'on lancer une pièce.

On découvrira à quoi servent vraiment les probabilités dans le chapitre sur les lois de probabilité. Savoir que quelque chose à une chance sur deux d'arriver, cela n'est absolument pas utile si on veut essayer l'expérience une fois. Que ce soit une probabilité d'une chance sur mille (1/1000), d'une chance sur deux (1/2) ou de neuf cent quatre-vingt-dix-neuf sur mille (999/1000), ça ne nous assurera pas du résultat futur.

Continuons dans nos exemples, et demandons nous si systématiquement les probabilités peuvent être trouvées par intuition, avec simplement un peu de bon sens.
